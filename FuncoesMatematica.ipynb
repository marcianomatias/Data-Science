```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

fig = plt.figure(figsize=(8,5))
plt.title("Função Constante")

x = np.linspace(-10,10,10000)
y = 20*np.ones(x.shape)

plt.plot(x,y)
plt.show()
```

Função Linear

```python
fig = plt.figure(figsize=(8,5))
plt.title("Função Linear")

x = np.linspace(-10,10,10000)
y = 5 +2*x

plt.plot(x,y)
plt.show()
```

Função Quadratica = f(x)= ax²+bx+c

```python
fig = plt.figure(figsize=(8,5))
plt.title("Função de segundo grau!")

x = np.linspace(0,5,10000)
y = 5 - 5*x**2

plt.plot(x,y)
plt.show()
```

Função de Terceiro grau

```python
fig = plt.figure(figsize=(8,5))
plt.title("Função de Terceiro grau")

x = np.linspace(-10,10,10000)
y = x**3 + 10*x**2 +5*x

plt.plot(x,y)
plt.show()
```

```python
fig = plt.figure(figsize=(8,5))
plt.title("Função de Racional")

x = np.linspace(0.1,0.5,10000)
y = (5e-15)*((1/x)**12 - (9.6/x)**6)

plt.plot(x,y)
plt.show()
```

Funções Trigonométricas

```python
fig = plt.figure(figsize=(8,5))
plt.title("Função de Seno")

x = np.linspace(-10,10,10000)
y = np.sin(x)

plt.plot(x,y)
plt.show()
```

```python
fig = plt.figure(figsize=(8,5))
plt.title("Função Cosseno")

x = np.linspace(-10,10,10000)
y = np.cos(x)

plt.plot(x,y)
plt.show()
```

Junção das 2 funções

```python
fig = plt.figure(figsize=(8,5))
plt.title("Função Cosseno")

x = np.linspace(-10,10,10000)
y = np.cos(x)

plt.plot(x,y)

x = np.linspace(-10,10,10000)
y = np.sin(x)

plt.plot(x,y)
plt.show()
```

```python
fig = plt.figure(figsize=(8,5))
plt.title("Função Tangente")

x = np.linspace(-1,1,10000)
y = np.tan(x)

plt.plot(x,y)
plt.show()
```

Função Exponencial

```python
fig = plt.figure(figsize=(8,5))
plt.title("Função Exponencial")

x = np.linspace(-10,10,10000)
y = np.exp(x)

plt.plot(x,y)
plt.show()
```

Função Logística

```python
fig = plt.figure(figsize=(8,5))
plt.title("Função Logística")

x = np.linspace(-10,10,10000)
y = 1/(1 + np.exp(-x))

plt.plot(x,y)
plt.show()
```
